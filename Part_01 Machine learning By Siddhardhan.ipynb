# Supervised Learning

In supervised Learning, the machine learning algorithm learns from labelled data

So, we already see that you know in machine learning the model learns from the data.

So what is meant by label data set lets say that machine learning model has to see an image and recognize whether the image 

represents an apple or mango.

What we will do is we take several images of apples and mangoes and we will tell the machine that these images belongs to 

apples and these are the images of mangoes.

Now these apples and mangoes are name it as labels and we feed the label dataset to our machine learning model.

Then our machine learning model or machine learning algorithm tries to find the pattern between these images and onces it 

has learned from the data when you give an unknown image it can correctly recognize whether the image represents an apple 

or mango.

So this is how supervised learning works and we are telling a supervision in terms of labels.

# Unsupervised Learning 

In Supervised learning ,the machine learning algorithm learns from unlabelled data.

Here we don't tell what that data represents,we won't give any labels .

Lets Consider a example

So we will give several images of apples and mangoes to our machine learning model and we won't tell that these images 

belong to apple or these images belong to mango

So we feed all these images without telling what it is to our machine learning model and what it does is. It tries to again 

find the pattern and it tries to group all these images and it will group the images into group one and group two

So all the apples will be grouped in one group and all the mangoes will be grouped in another group .

When you give a new image of an apple or a mango it tells you whether it belongs to group one or group two .

This is called as supervised learning because we are not giving any supervision interms of labels

# Reinforcement Learning

Reinforcement learning is not similar to supervised or unsupervised learning.

It's quite different from both other types

Lets try to understand this in more detail

Reinforcement learning is an area of machine learning concerned with how intelligent agents take actions in an environment 

to maximize its rewards

There are four main aspects in reinforcement learning.They are

1.Environment

2.Agent

3.Action

4.Reward

So, there will be an environment and we need to build an agent that acts in that environment .

So,that agent in that environment tries to take some actions and for that action it agains some rewards.

For Example

we want to make a computer software or a computer program that can play chess  like a human being

Here our chess board becomes the enviornment and our computer become the agent.

In the environment of chessboard our agent which is the computer tries to take actions.

The actions represents the move the computer takes in the chess and for each step it gets a reward .

So the ultimate reward is winning the chess games and for each step it takes closer to winning it will get a positive 

reward.

If it takes a bad step or bad move,so it will get a negative reward.

By this machine tries to learn how to play that game.

Several applications are there for reinforcement learning and all autonomous systems like cars ,Automatic drones are based 

on reinforcement learning.

we will try to make an agent that acts in an environment to increase its chance of winning.

***an "agent" refers to the entity or system that interacts with an environment in order to learn how to perform tasks and maximize certain objectives. The agent's goal is to take actions within the environment to achieve a desired outcome or maximize a cumulative reward over time.

**the "environment" refers to the external system or context in which an agent operates and learns. The environment encompasses everything that the agent interacts with and from which it receives feedback. It provides the context in which the agent takes actions and receives rewards based on its behavior.
